## 워드클라우드 전처리

```python
import warnings
warnings.filterwarnings("ignore") # 경고 무시

import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('./namuwiki.csv')
df.head()
```

```python
st='한글만 추출하기 English 韓字!!!'
hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자
# [^내용]: 내용에 일치하지 않는 문자 검색

# 한글과 띄어쓰기를 제외한 모든 부분을 제거
result = hangul.sub('', st) 
print(result)

# 한글과 띄어쓰기를 제외한 모든 부분만 찾음
result = hangul.findall(st) 
print(result)
```

```python
def cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자
    # 한글과 띄어쓰기를 제외한 모든 부분을 제거
    result = hangul.sub('', str(text))
    return result
```

```python
cleaning(df['content_text'][0])
```

```python
# title, category, content_text
# lambda 파라미터: 함수 body
df['title'] = df['title'].apply(lambda x: cleaning(x))
df['category'] = df['category'].apply(lambda x: cleaning(x))
df['content_text'] = df['content_text'].apply(lambda x: cleaning(x)) # 왜지?
```

```python
df.head()
```

```python
# 말뭉치(corpus, 모든 한글의 통합)
print(type(df['title'])) # pandas.core.series.Series
print(type(df['title'].tolist())) # df를 리스트로 변환

"".join(df['title'].tolist()) # 순수 한글만 추출해서 통문자열로 만듦
```

```python
# 말뭉치(corpus, 모든 한글의 통합): Series -> list -> str
title_corpus = "".join(df['title'].tolist())
category_corpus = "".join(df['category'].tolist())
content_corpus = "".join(df['content_text'].tolist())
print(title_corpus)
```

```python
from konlpy.tag import Okt # Okt tagger의 사용
from collections import Counter

nouns_tagger = Okt() 

# konlpy의 형태소 분석기로 명사 단위의 키워드를 추출
nouns = nouns_tagger.nouns(content_corpus) # 의미 기반으로 접근
count = Counter(nouns)
```

```python
# 하나의 글자로 구성된 명사는 삭제
# 문자 길이가 1보다 작은 값을 제거
rm_char_counter = Counter({x:count[x] for x in count if len(x) > 1})
rm_char_counter
```

```python
# 불용어(의미가 부족한 단어 제거)
# with open: 파일 객체 사용 후 자동 close
# f에 파일 객체 할당
with open('./Korean_stopwords.txt', encoding='utf-8') as f:
    stopwords = f.readlines()

# 혹시 모르기 때문에 문자열 주변 공백을 제거할 필요가 있음
print(type(stopwords)) # <class 'list'>
```

```python
# 문자열 주변 공백 제거
stopwords = [ x.strip() for x in stopwords ]
print(stopwords[:10])
```

```python
# 나무위키 페이지에 등장하는 불용어를 추가
namu_wiki_stopwords = ['상위', '문서', '내용', '누설', '아래', '해당', '설명', '표기', '추가', '모든', '사용', '매우', '가장',
                       '줄거리', '요소', '상황', '편집', '틀', '경우', '때문', '모습', '정도', '이후', '사실', '생각', '인물', 
                       '이름', '년월']
for stopword in namu_wiki_stopwords:
    stopwords.append(stopword)
```

```python
print('불용어수: ', len(stopwords))
```

```python
# 말뭉치에서 불용어 제거
# stopwords에 포함되지 않는 것들만 추출, stopwords는 제거
content_counter = Counter({x:count[x] for x in rm_char_counter if x not in stopwords})
```

```python
content_counter
```

## 워드클라우드

```python
import random
import pytagcloud
import webbrowser

# 가장 출현 빈도수가 높은 60개의 단어를 선정합니다.
ranked_tags = content_counter.most_common(60)

# pytagcloud로 출력할 60개의 단어를 입력합니다. 단어 출력의 최대 크기는 80으로 제한합니다.
taglist = pytagcloud.make_tags(ranked_tags, maxsize=80)

# pytagcloud 이미지를 생성합니다. 폰트는 나눔 고딕을 사용합니다.
# rectangular: 사각형 모양의 형태
pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(800, 500), fontname='NanumGothic', rectangular=False)

# 생성한 이미지를 주피터 노트북상에서 출력
from IPython.display import Image
Image(filename='wordcloud.jpg')
```

```python
nouns_tagger = Okt() 

# konlpy의 형태소 분석기로 명사 단위의 키워드를 추출
nouns = nouns_tagger.nouns(content_corpus) # 의미 기반으로 접근
count = Counter(nouns)

# 하나의 글자로 구성된 명사는 삭제
# 문자 길이가 1보다 작은 값을 제거
rm_char_counter = Counter({x:count[x] for x in count if len(x) > 1})

# 말뭉치에서 불용어 제거
# stopwords에 포함되지 않는 것들만 추출, stopwords는 제거
content_counter = Counter({x:count[x] for x in rm_char_counter if x not in stopwords})

# 가장 출현 빈도수가 높은 60개의 단어를 선정합니다.
ranked_tags = content_counter.most_common(60)

# pytagcloud로 출력할 60개의 단어를 입력합니다. 단어 출력의 최대 크기는 80으로 제한합니다.
taglist = pytagcloud.make_tags(ranked_tags, maxsize=50)

# pytagcloud 이미지를 생성합니다. 폰트는 나눔 고딕을 사용합니다.
pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(800, 500), fontname='NanumGothic', rectangular=True)

# 생성한 이미지를 주피터 노트북상에서 출력
Image(filename='wordcloud.jpg')
```
