### Google 지도 API를 활용한 장소 검색
### Naver, Daum API는 기업만 지원으로 변경됨.

```python
import googlemaps

def find_places(searching):
    my_key = "구글 맵 api key"

    # 구글맵스 객체 생성하기
    maps = googlemaps.Client(key=my_key)

    # 지오코딩 API 결과값 호출하여 geo_location 변수에 저장
    geo_location = maps.geocode(searching, language = 'ko')[0].get('geometry')
    y = geo_location['location']['lat']
    x = geo_location['location']['lng']
    print("위도:", y)
    print("경도:", x)
    data = [searching, x, y, searching]
    return data
```

```python
data = find_places('제주공항')
data
```

```python
from tqdm.notebook import tqdm  # version 4.46 이후 사용시 설정
```

```python
print(len(locations))
```

```python
# 인스타그램 위치명 위치정보 검색하기
# 검색시 시간이 많이 걸려, 현재 작업 진행현황을 볼 수 있도록 처리
import time
locations_inform = []
cnt = 0
for location in tqdm(locations):
    try:
        # location based system
        lbs = find_places(location)
        locations_inform.append(lbs)
        time.sleep(0.5)
        cnt += 1
        if cnt == 5:
            break
    except:
        pass

locations_inform
```

```python
# 인스타 게시량 및 위치정보 데이터 불러오기
# index_col = 0: 첫번째 엑셀 칼럼
import pandas as pd

location_counts_df = pd.read_excel('./data/location_counts.xlsx', index_col = 0) # 게시글 빈도수
location_counts_df.head()
```

```python
# 위치 정보 좌표
locations_inform_df = pd.read_excel('./data/locations.xlsx')
locations_inform_df.head()
```

```python
# 위치 데이터 병합하기
# name_official과 오른쪽 테이블 인덱스가 일치하면 merge -> 인스타위치명을 사용하지 않는 이유는 식당이름이 아닌 것을 걸러내기 위해?
# right_index=True: 병합시 오른쪽 DataFrame은 index 값을 사용함.
location_data = pd.merge(locations_inform_df, location_counts_df, 
                         how = 'inner',
                         left_on = 'name_official',
                         right_index=True)

location_data.head()
```

```python
# 추가) 동일한 name_official 가지는 데이터 --> 인스타 게시 횟수 합치기
# index = ['name_official','경도','위도']: 3개의 컬럼을 그룹화
# values = 'place': 빈도수
# aggfunc='sum': 합
location_data = location_data.pivot_table(
    index = ['name_official','경도','위도'], 
    values = 'place', 
    aggfunc='sum')
location_data.head(10)
```

```python
# 병합한 데이터 저장하기
location_data.to_excel('./data/location_inform.xlsx')
```

```python
# 데이터 불러오기
location_data = pd.read_excel('./data/location_inform.xlsx')
location_data.info()
```

```python
# 지도 표시하기
import folium

Mt_Hanla =[33.362500, 126.533694]  # 지도 중심
# zoom_start = 11: 1 ~ 17 단계
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11) 

for i in range(len(location_data)):
    name = location_data ['name_official'][i]    # 공식명칭
    count = location_data ['place'][i]           # 게시글 개수
    size = int(count) * 1  # 마커의 크기를 확대함
    lat = float(location_data['위도'][i])      
    long = float(location_data['경도'][i])  
    # 마커 출력
    folium.CircleMarker((lat,long), radius = size, weight=5, color='red', popup=name).add_to(map_jeju)
    
map_jeju
```

```python
# 지도 저장하기
map_jeju.save('./data/jeju.html')
```

```python
# 지도 그룹으로 표시하기
from folium.plugins import MarkerCluster

locations = [] # 좌표
names = []     # 장소 이름 

for i in range(len(location_data)):
    data = location_data.iloc[i]  # 행 하나씩
    # 위도, 경도를 튜플로 저장
    locations.append((float(data['위도']),float(data['경도'])))
    names.append(data['name_official'])


# 자바스크립트 소스
icon_create_function = """\
function(cluster) {
    return L.divIcon({
    html: '<b>' + cluster.getChildCount() + '</b>',
    className: 'marker-cluster marker-cluster-large',
    iconSize: new L.Point(30, 30)
    });
}"""
```

```python
Mt_Hanla =[33.362500, 126.533694]  # map center
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start = 11)
                       
marker_cluster = MarkerCluster(
    locations=locations,
    popups=names,
    name='Jeju',
    overlay=True,
    control=True,
    icon_create_function=icon_create_function
)

marker_cluster.add_to(map_jeju2)  # 마커에 지도 할당
folium.LayerControl().add_to(map_jeju2) # folium에 지도 할당

map_jeju2
```
