```python
import pandas as pd
```

```python
# 엑셀 읽어오기
raw = pd.read_excel('./data/crawling_raw.xlsx')
```

```python
raw['tags'][:3]
```

```python
tags = raw['tags'][0]
tags = tags[1:-1]
tags_list = tags.split(', ')

for tag in tags_list:
    print(tag)
```

```python
# ''까지 제거
import re
tmp = re.sub('\'', '', tags)
tmp_list = tmp.split(', ')

for tag in tmp_list:
    print(tag)
```

```python
# 해시태그의 통합
tags_total = []
for tags in raw['tags']:
    # print(tags) # ['#제주분식', ...]
    # print(type(tags)) # 리스트 아님, <class 'str'>
    tags = tags[1:-1] # [ ]를 정규식으로 제거
    tags = re.sub('\'', '', tags) # ' ' 제거
    tags_list = tags.split(', ')
    
    for tag in tags_list:
        tags_total.append(tag)
tags_total[:5]
```

```python
# 빈도분석
from collections import Counter
```

```python
tags_counts = Counter(tags_total)
# type(tags_counts) # collections.Counter
```

```python
# 가장 자주 나오는 태그 5개 출력
tags_counts.most_common(5)
```

```python
# 데이터 정제하기
STOPWORDS = ['#일상', '#선팔', '#제주도', '#jeju', '#반영구', '#제주자연눈썹',
 '#서귀포눈썹문신', '#제주눈썹문신', '#소통', '#맞팔',
 '#제주도맛집', '#제주맛집', '#제주관광', '#제주여행',
 '#제주도여행', '#제주', '#서귀포맛집', '#제주도가볼만한곳',
 '#제주도핫플레이스', '#제주핫플레이스', '#서귀포', '#제주맛집추천']
```

```python
tags_total_selected = []
for tag in tags_total:
    if tag not in STOPWORDS and tag != '':
        tags_total_selected.append(tag)
tags_counts_selected = Counter(tags_total_selected)
tags_counts_selected.most_common(5)
```

```python
# 시각화 라이브러리 호출 및 환경 설정(한글 폰트)
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import sys

if sys.platform in ["win32", "win64"]: # Windows
    font_name = "malgun gothic"
elif sys.platform == "darwin": # Mac
    font_name = "AppleGothic"

rc('font',family=font_name)
%matplotlib inline
```

```python
df = pd.DataFrame(tags_counts_selected.most_common(30))
df.columns = ['tags', 'counts'] # 칼럼명 설정 방법
df.head()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='counts', y='tags', data=df)
plt.show()
```

```python
from wordcloud import WordCloud

# 폰트 지정
import platform
if platform.system() == 'Windows':   # 윈도우의 경우
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == "Darwin":   # Mac 의 경우
    font_path = "/Users/$USER/Library/Fonts/AppleGothic.ttf"
```

```python
# 워드클라우드 만들기
# background_color="white": 배경색
# max_words=100: 최대 출력할 단어의 수
# relative_scaling= 0.3: 빈도에 따라 상대적인 크기
# width = 800: 너비
# height = 400: 높이
                    
wordcloud=WordCloud(font_path= font_path, 
                    background_color="white",
                    max_words=100,
                    relative_scaling= 0.1,
                    width = 800,
                    height = 400
                 ).generate_from_frequencies(tags_counts_selected) # Counter 객체를 파라미터로 넘김
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./files/tag-wordcloud.png')
```
