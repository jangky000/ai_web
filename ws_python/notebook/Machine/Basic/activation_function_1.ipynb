```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
# windows 10
# font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgunsl.ttf").get_name()
rc('font', family=font_name)           # 맑은 고딕 폰트 지정
plt.rcParams["font.size"] = 12         # 글자 크기
# plt.rcParams["figure.figsize"] = (10, 4) # 10:4의 그래프 비율
plt.rcParams['axes.unicode_minus'] = False  # minus 부호는 unicode 적용시 한글이 깨짐으로 설정

# Jupyter에게 matplotlib 그래프를 출력 영역에 표시할 것을 지시하는 명령
%matplotlib inline  
```

```python
# 계단 함수
def step_function(x):
    return np.where(x < 0, 0, 1)
    # return np.where(x < 0, -1, 1)
    
# 시작점과 끝점을 균일 간격으로 나눈 점들을 생성해주는 linspace()함수
x = np.linspace(-5, 5, 50) # 50개 구간으로 50개의 값 생성
print(x)
y = step_function(x) # 초창기 스텝 함수 -> 컴퓨터의 성질
print(y)
plt.plot(x, y)
plt.show()
```

```python
# Sigmoid 함수: 이항 분류, 참/거짓, 합격/불합격, 성공/실패
# 큰값이 올수록 명확해짐 :      -25 ~ 25
# 값이 작을수록 판단이 어려움: -1 ~ +1
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x)) # e: 2.71828...

x = np.linspace(-25, 25, 50)
print(x)
y = sigmoid_function(x)

plt.plot(x, y)
plt.show()
```

```python
# ReLU 함수: 은닉층에서의 유효한 값 유지
# 연산 단순, 속도 빠름
#: 음수는 버리고, 양수는 그대로 출력
def relu_function(x):
    return np.where(x <= 0, 0, x)

x = np.linspace(-5, 5, 50)
print(x)
y = relu_function(x)

plt.plot(x, y)
plt.show()
```

```python
# 항등 함수: 수치예측 선형회귀에서 계산된값을 그대로 출력
x = np.linspace(-5, 5, 50)
print(x)
y = x

plt.plot(x, y)
plt.show()
```

```python
# Softmax 함수: 다중 분류에서 많이 사용, 사물 인식, 품종 분류
# 전체 y의 합이 1
# 수의 비중이 큰 수가 확률이 더 높게 나옴 -> 큰 수는 더 크게 작은 수는 더 작게
def softmax_function(x):
    return np.exp(x) / np.sum(np.exp(x))

x = np.array([1, 2, 3])
print(x)
y = softmax_function(x)

print(y)
print('Total:', np.sum(y))
plt.plot(x, y)
plt.show()
```

```python
# 하이퍼볼릭 탄젠트, RNN 등 순환 신경망에 주로 사용
# 언어의 조합, 시계열 자료 예측
def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5, 5, 50)
y = tanh_function(x)

plt.plot(x, y)
plt.show()
```
