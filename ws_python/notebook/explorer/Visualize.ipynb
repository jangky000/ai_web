```python
# seaborn에서 iris 데이터를 가져옴
df = sns.load_dataset("iris") 
print(df.head(10))
print(type(df))
```

```python
df.plot()
# plt.figure(figsize=(8, 3))  # 작동안됨
# plt.rcParams["figure.figsize"] = (10, 4) # 10:4의 그래프 비율, plot 출력전에 선언 
plt.title("붖꽃 데이터의 분포")
plt.xlabel("갯수")
plt.ylabel("크기(cm)")
plt.show()
# 초록색은 분석을 피하는 것이 좋다. 확실하게 구분되는 값을 분석하는 것이 좋다.
```

```python
df.hist()
plt.show()
# 균등 분포는 거의 존재하지 않는다 and 학습할 이유도 없다.
```

```python
# kind = bar: 수직 막대 그래프
df.sepal_length[:20].plot(kind='bar', rot=0) # rot=0: x ticks 회전 각도
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
# 하나의 꽃 별로 데이터를 바 형태로 출력
df[:10].plot.bar(rot=0)
plt.title("Iris 데이터의 Bar Plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 7)
plt.show()
```

```python
# 종 별로 그룹화해서 평균을 구함
df2 = df.groupby(df.species).mean() # 열 평균
df2.columns.name = "feature"
df2
```

```python
# 종별로 시각화
df2.plot.bar(rot=0)
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("종류")
plt.ylim(0, 8)
plt.show()
```

```python
# T: 전치 행렬
df2.T.plot.bar(rot=0)
plt.title("각 Feature의 종 별 평균")
plt.xlabel("종류")
plt.ylabel("평균")
plt.show()
```

```python
df.plot.hist()
plt.show()
```

```python
df.plot.kde()  # 커널 밀도: 히스토그램의 적분 기반 면적을 1로 데이터를 표시한 그래프
plt.title("각 Feature 값들의 빈도수에 대한 Kernel Density Plot")
plt.xlabel("데이터 값")
plt.show()
```

```python
# 이상치 확인 -> 이상치 제거 -> IQR을 확인 후에 제거
df.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
df.boxplot(by='species')
plt.tight_layout(pad=2, h_pad=1)
plt.show()
```

```python
df.plot.scatter(x='petal_length', 
                y='petal_width', 
                s=50, # marker size
                c='blue', 
                alpha=0.5)
plt.title('Scatter Plot of iris by pandas', fontsize=20)
plt.xlabel('Petal Length', fontsize=14)
plt.ylabel('Petal Width', fontsize=14)
plt.show()
```

```python
# species의 종류만큼 마커를 생성할 것
# seaborn: pair plot 상관 관계 확인 -> 상관성이 높은 데이터를 기준으로 분석하는 것이 좋다.
sns.pairplot(df, hue="species", markers=["o", "s", "D"]) # 종별로, 동그라미, 네모, 다이아몬드
plt.title("Iris Data의 Pair Plot")
plt.show()
```

### 타이타닉 데이터

```python
# kaggle에 가면 더 많고 다양한 데이터가 있다. 거의 모든
```

```python
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
titanic.head()
```

```python
# 등석에 따른 비율
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```
