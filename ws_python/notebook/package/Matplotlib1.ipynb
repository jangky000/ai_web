```python
import matplotlib.pyplot as plt
%matplotlib inline  

train_loss = [10.3, 9.7, 8.5, 7.7, 6.2, 3.3, 2.5, 1.2]
val_loss = [8.5, 5.7, 4.5, 3.7, 2.3, 1.5, 1.3, 0.2]
train_accuracy = [0.1, 0.25, 0.45, 0.57, 0.70, 0.81, 0.95, 1.0]
val_accuracy = [0.05, 0.10, 0.25, 0.31, 0.41, 0.49, 0.75, 0.91]

fig, loss_ax = plt.subplots()
# plt.figure(figsize=(6,4)) # ERROR
fig.set_size_inches(10, 5)  # 챠트 크기 설정

# 왼쪽 y 축 설정
loss_ax.plot(train_loss, 'y', label='train loss')
loss_ax.plot(val_loss, 'r', label='val loss')
loss_ax.set_ylim([0.0, max(y1)+5]) # 값을 반영하여 변경

acc_ax = loss_ax.twinx()  # x 축 공유 설정

# 오른쪽 y 축 설정
acc_ax.plot(train_accuracy, 'b', label='train accuracy')
acc_ax.plot(val_accuracy, 'g', label='val accuracy')
acc_ax.set_ylim([0.0, 1]) # 0.0, 1: 0 ~ 100 %, 정확도임으로 변경하지 않음

# 축 레이블 설정
loss_ax.set_xlabel('epoch')  # x 축 레이블 설정, 학습 횟수
loss_ax.set_ylabel('loss')   # 왼쪽 y 축 레이블 설정, 오차

acc_ax.set_ylabel('accuracy') # 오른쪽 y 축 레이블 설정, 정확도

loss_ax.legend(loc='upper left') # 오차 레이블 위치
acc_ax.legend(loc='lower left')  # 정확도 레이블 위치

plt.show()
```

```python
# 기상관측된 데이터 전체를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
f = open('./seoul.csv', 'r', encoding='utf-8') # r: read 읽기모드
data = csv.reader(f, delimiter=',') # delimiter: 구분자
header = next(data) # 현재 라인을 읽고, 파일 지시자를 다음 라인으로 이동(첫줄읽음)
high = []


for row in data:
    if row[-1] != '': # 데이터가 있는 경우
        high.append(float(row[-1]))

        
plt.hist(high, bins = 1000, color='tomato', label='최고온도') # c는 에러 발생, tomato 오타나도 에러 발생-> 근데 high가 틀렸다고 나옴
plt.title('서울 지역 최고온도')
plt.xlabel('온도')
plt.ylabel('횟수')
plt.legend()
plt.grid(True)
plt.show()
        
f.close()
```

```python
# 기상관측된 데이터중 1월 데이터를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
f = open('./seoul.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
header = next(data)

high = []

for row in data:
    if row[-1] == '':
        continue
    month = row[0].split('-')[1]
    if month=='01':
        high.append(float(row[-1]))
#         print(row)
#         if len(high) == 5:
#             break

plt.hist(high, bins=100, color='lightseagreen', label='최고온도')
plt.xlabel('온도')
plt.ylabel('횟수')
plt.legend()
plt.grid(True)
plt.show()        
f.close()
```

```python
# 기상관측된 데이터중 8월 데이터를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
f = open('./seoul.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
header = next(data)

high = []

for row in data:
    if row[-1] == '':
        continue
    month = row[0].split('-')[1]
    if month=='08':
        high.append(float(row[-1]))
#         print(row)
#         if len(high) == 5:
#             break

plt.hist(high, bins=100, color='fuchsia', label='최고온도')
plt.xlabel('온도')
plt.ylabel('횟수')
plt.legend()
plt.grid(True)
plt.show()        
f.close()
```

```python
# 기상관측된 데이터중 1월, 8월 데이터를 대상으로하여 서울 지역의 최고 온도를
# histogram으로 출력할것, 구간갯수: 100.
# 하나의 좌표에 출력
f = open('./seoul.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
header = next(data)
jan = [] # 1월
aug = [] # 8월

for row in data:
    if row[-1] != '':
        date_split = row[0].split('-')
        month = date_split[1]
        if month == '01':
            jan.append(float(row[-1])) # 최고 기온
        elif month == '08':
            aug.append(float(row[-1])) # 최고 기온

# y: 갯수, x: 데이터, bins: 값의 구간의 갯수 
plt.hist(jan, bins=100, color='lightseagreen', label='1월 최고 온도') 
plt.hist(aug, bins=100, color='fuchsia', label='8월 최고 온도')  
plt.title('서울 지역의 1월, 8월 최고 온도')
plt.xlabel('온도')
plt.ylabel('횟수')
plt.legend() # 범례
plt.grid(True)
plt.show()

f.close()
```

## Boxplot

```python
# 1월달의 최고기온을 대상으로 boxplot 출력
f = open('./seoul.csv', encoding='utf-8')
data = csv.reader(f)
next(data) # header skip
jan = []

for row in data:
    if row[-1] != '':  # ['1907-10-01', '108', '13.5', '7.9', '20.7']
        row_split = row[0].split('-') # 2018-12-25
        month = row_split[1] 
        if month == '01':
            jan.append(float(row[-1]))

# plt.style.use('ggplot')
plt.boxplot(jan) # 이상치 존재, Outlier가 많다
plt.title('1월달의 최고기온')
plt.grid(True)
plt.show()
f.close()
```

```python
# boxplot 여러 개 출력
# 1월, 8월달의 최고기온을 대상으로 boxplot 출력
f = open('./seoul.csv', encoding='utf-8')
data = csv.reader(f)
next(data) # header skip
jan = []
aug = []

for row in data:
    if row[-1] != '':  # ['1907-10-01', '108', '13.5', '7.9', '20.7']
        row_split = row[0].split('-') # 2018-12-25
        month = row_split[1] 
        if month == '01':
            jan.append(float(row[-1]))
        elif (month == '08'):
            aug.append(float(row[-1]))

plt.style.use('ggplot')
plt.boxplot([jan, aug]) # list로 지정
plt.title('1월, 8월달의 최고기온')
plt.grid(True)
plt.show()
f.close()
```
