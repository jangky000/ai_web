```python
import warnings
warnings.filterwarnings("ignore") # 경고 무시

import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# with open('./data.txt') as f:
#     data = f.readlines()
# data

df = pd.read_csv('./data.txt', sep='\n', encoding='utf-8', header=None)
df
```

```python
# 2. Text 파일을 읽어 한글만 추출하는 코드를 함수로 작성하시오.
def cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자
    # 한글과 띄어쓰기를 제외한 모든 부분을 제거
    result = hangul.sub('', str(text))
    return result
```

```python
# 3. Lambda 함수를 이용하여 한글만 추출하는 함수를 호출하는 코드를 작성하시오.
df[0] = df[0].apply(lambda x: cleaning(x))
# 4. 말뭉치(corpus, 모든 한글의 통합)를 생성하는 코드를 작성하시오.
corpus = " ".join(df[0].tolist())
corpus
```

```python
from konlpy.tag import Okt # Okt tagger의 사용
from collections import Counter

nouns_tagger = Okt() 

# 5. konlpy의 형태소 분석기로 명사 단위의 키워드를 추출 코드를 제작하시오.
nouns = nouns_tagger.nouns(corpus)
# 6. Counter 모듈을 사용하여 명사별 카운트를 처리한 객체를 생성하는 코드를 작성하시오.
count = Counter(nouns)
```

```python
count
```

```python
# 7. 하나의 글자로 구성된 명사를 삭제하는 코드를 작성하시오.
rm_char_counter = Counter({x:count[x] for x in count if len(x) > 1})
rm_char_counter
```

```python
with open('./Korean_stopwords.txt', encoding='utf-8') as f:
    stopwords = f.readlines()

# 8. 문자열 주변 공백을 제거하는 코드를 작성하시오.
stopwords = [ x.strip() for x in stopwords ]
```

```python
# 나무위키 페이지에 등장하는 불용어를 추가
news_stopwords = ['이데일리', '입력', '폰트', '변경', '크게', '메일', '인쇄', '목록', '고객', '센터', '이동', '기자', '해도', '연합뉴스', '보기', '대한', '하자', '통해', '등록', '전만', '정도', '보고', '사진', '뉴시스', '지난', '대해', '지금', '올해', '여러', '경우', '라면', '그동안', '라며', '연합뉴스', '위해', '별로', '대로', '오전', '억원', '미만', '중이', '만큼', '']
for stopword in news_stopwords:
    stopwords.append(stopword)
```

```python
# 9. 말뭉치에서 불용어를 제거하는 코드를 작성하시오.
content_counter = Counter({x:count[x] for x in rm_char_counter if x not in stopwords})
```

```python
content_counter
```

```python
# 10. 워드 클라우드 이미지를 생성하고 파일로 저장 한 후 이미지를 첨부하시오.
import random
import pytagcloud
import webbrowser

# 가장 출현 빈도수가 높은 60개의 단어를 선정합니다.
ranked_tags = content_counter.most_common(60)

# pytagcloud로 출력할 60개의 단어를 입력합니다. 단어 출력의 최대 크기는 80으로 제한합니다.
taglist = pytagcloud.make_tags(ranked_tags, maxsize=80)

# pytagcloud 이미지를 생성합니다. 폰트는 나눔 고딕을 사용합니다.
# rectangular: 사각형 모양의 형태
pytagcloud.create_tag_image(taglist, 'news_wordcloud.jpg', size=(800, 500), fontname='NanumGothic', rectangular=False)

# 생성한 이미지를 주피터 노트북상에서 출력
from IPython.display import Image
Image(filename='news_wordcloud.jpg')
```
