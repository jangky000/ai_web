```python
# 라이브러리 임포트
import pandas as pd
```

```python
# 시군구 목록 데이터 불러오기
seoul_sgg = pd.read_excel('./data/seoul_sgg_202005.xlsx')
seoul_sgg.head()
```

```python
# 서울시 스타벅스 매장 목록 데이터 불러오기
seoul_starbucks = pd.read_excel('./data/seoul_starbucks_gu_202005.xlsx')                                 
seoul_starbucks.head()
```

```python
# 시군구별 스타벅스 매장 수 세기
# index: 그룹 컬럼, values: 값 컬럼, aggfunc: 함수
starbucks_sgg_count = seoul_starbucks.pivot_table(
                                                index = '시군구명', 
                                                values='매장명', 
                                                aggfunc='count'
                                                ).rename(columns={'매장명':'스타벅스_매장수'})
starbucks_sgg_count.head()
```

```python
# 서울시 시군구 목록 데이터에 스타벅스 매장 수 데이터를 병합
# how='left': 왼쪽 데이터 프레임은 모두 출력됨
# 한번만 실행 -> 두번 이상 실행하면 계속 칼럼이 추가됨
seoul_sgg = pd.merge(seoul_sgg, starbucks_sgg_count, how='left', on='시군구명')
seoul_sgg
```

```python
# 서울시 시군구별 인구통계 데이터 불러오기
seoul_sgg_pop = pd.read_excel('./data/sgg_pop_202005.xlsx')
seoul_sgg_pop.head()
```

```python
# 서울시 시군구 목록 데이터에 서울시 시군구별 인구통계 데이터를 병합
# seoul_sgg: 스타벅스 매장 수가 구 별로 들어가 있음
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_pop, how='left', on='시군구명')
seoul_sgg
```

```python
# 서울시 시군구 목록 데이터에 서울시 시군구별 사업체 수 통계 데이터를 병합
# seoul_sgg: 스타벅스 구 별 매장 수 + 주민등록인구수
seoul_sgg_biz = pd.read_excel('./data/sgg_biz_202005.xlsx')
seoul_sgg = pd.merge(
    seoul_sgg, 
    seoul_sgg_biz,
    how='left',
    on='시군구명'
)
seoul_sgg
```

```python
# 병합 결과를 엑셀 파일로 저장
seoul_sgg.to_excel('./data/seoul_sgg_stat_202005.xlsx', index=False)
```
