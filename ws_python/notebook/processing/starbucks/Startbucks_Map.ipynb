```python
# 라이브러리 임포트
import pandas as pd
import folium # 지도 시각화 패키지, pip install folium
import json
```

```python
sbs = pd.read_excel('./data/seoul_starbucks_202005.xlsx')
sbs.head()
```

```python
sbs['매장타입'].unique()
```

```python
# folium을 이용한 지도 생성, zoom_start: 1 ~ 18 지원
# location; 지도 중심 좌표, Google Map등에서 확인 가능 # 종로구청 위치
sbs_map = folium.Map(
    location=[37.573050, 126.979189], tiles ='CartoDB positron' , zoom_start=11 )

# 위치에 마커 추가하기
for idx in sbs.index: # 0~516
    lat = sbs.loc[idx, '위도']
    lng = sbs.loc[idx, '경도']
    store_type = sbs.loc[idx, '매장타입']
#     print(idx, lat, lng)

    # 원으로 마커 표시
    folium.CircleMarker(
        location=[lat, lng],
        fill=True, # 색을 칠한다.
        fill_color = 'green',
        fill_opacity=1.0, # 진한 색상
        color='yellow', # 테두리 색
        weight=1, # 테두리가 진해짐
        radius=3 # 표시되는 원의 크기
    ).add_to(sbs_map)
    
#folium은 html 저장을 지원함
sbs_map.save('./output/sbs_map.html')
sbs_map
```

```python
# folium을 이용한 지도 생성, zoom_start: 1 ~ 18 지원
# location; 지도 중심 좌표, Google Map등에서 확인 가능 # 종로구청 위치
sbs_map2 = folium.Map(
    location=[37.573050, 126.979189], tiles ='CartoDB positron' , zoom_start=11 )

for idx in sbs.index: # 0~516
    lat = sbs.loc[idx, '위도']
    lng = sbs.loc[idx, '경도']
    store_type = sbs.loc[idx, '매장타입']
#     print(idx, lat, lng)

    fcolor = ''
    size = 0
    if store_type == 'general':
        fcolor='gray'
        size = 3
    elif store_type == 'reserve':
        fcolor = 'seagreen'
        size = 5
    elif store_type == 'generalDT':
        fcolor = 'tomato'
        size = 5
    # 원으로 마커 표시
    folium.CircleMarker(
        location=[lat, lng],
        fill=True, # 색을 칠한다.
        fill_color = fcolor,
        fill_opacity=1.0, # 진한 색상
        color='yellow', # 테두리 색
        weight=1, # 테두리가 진해짐
        radius=size # 표시되는 원의 크기
    ).add_to(sbs_map2)
    
sbs_map2.save('./output/sbs_map2.html')
sbs_map2
```
