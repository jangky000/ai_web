```python
import pandas as pd
```

```python
# kto_201901.xlsx 엑셀 데이터 불러오기
'''
header=0: 1번째 줄이 변수명
header=1: 2번째 줄이 변수명
usecols='A:G': 사용할 엑셀의 컬럼
skipfooter=4: 마지막 4개의 라인 skip -> 집계
'''
kto = pd.read_excel('./data/kto_201901.xlsx', 
                          header=1,
                          usecols='A:G',
                          skipfooter=4)
kto.head()
```

```python
# 마지막 5개의 행 출력
kto.tail()
```

```python
kto.info()
#  67 entries: 레코드 개수
# total 7 columns: 변수 개수
```

```python
# 기술 통계
kto.describe()
# min에 0이 나오고 있다.
```

```python
# 각 칼럼에서 0을 필터링
# \는 라인의 연결을 뜻함, 라인 계속, 
condition = (kto['관광'] == 0) | (kto['상용'] == 0) \
        | (kto['공용'] == 0) | (kto['유학/연수'] == 0)
# condition
# 0이 있는 데이터를 찾아냄
kto[condition] # true인 행만 필터링해서 출력
```

```python
# 날짜 칼럼 추가
kto['기준년월'] = '2019-01'
kto.head()
```

```python
# 국가명을 고유하게 출력
kto['국적'].unique()
```

```python
# 대륙목록 만들기
# 구주: 서유럽, 대양주: 오세아니아(호주 등)
continents = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
continents
```

```python
# 순수 국가명만 추출
# 대륙 목록을 제거
condition = (kto.국적.isin(continents) == False) # False ==  False이면 True -> 즉 목록 안에 없는 것들을 추출
kto_country = kto[condition]
kto_country['국적'].unique()
```

```python
kto_country.head(30) # index 26번이 없음, 대륙명 등을 제거했기 때문
```

```python
# 인덱스 다시 설정
kto2 = kto_country.reset_index(drop=True)
kto2.head(30)
```

```python
# 출력되는 국가명이 대륙 순서순으로 되어 있음
# 대륙 컬럼값 만들기
continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 \
+ ['아프리카']*2 + ['기타대륙'] + ['교포']
print(continents)
```

```python
kto2['대륙'] = continents
kto2.head(30)
```

```python
# 데이터프레임의 값 접근하기
print('len: ', len(kto)) # 67
continents = [] # 대륙명 저장

# for row in kto:
#     print(row) # 칼럼명만 출력

for i in range(len(kto)):
    nation_name = kto.loc[i][0]
#     print(nation_name, end=', ')
    if nation_name in (['일본', '대만', '홍콩', '마카오', '태국', \
                         '말레이시아', '필리핀', '인도네시아', \
                         '싱가포르','미얀마', '베트남', '인도', \
                         '스리랑카', '파키스탄', '방글라데시', \
                         '캄보디아', '몽골', '중국','이란', \
                         '이스라엘', '터키', '우즈베키스탄', \
                         '카자흐스탄', 'GCC', '아시아 기타']):
        continents.append('아시아')
    elif nation_name in (['미국', '캐나다', '멕시코', '브라질', \
                           '미주 기타']):
        continents.append('아메리카')
    elif nation_name in (['영국', '독일', '프랑스', '네덜란드', \
                           '스웨덴', '스위스', '이탈리아', '덴마크', \
                           '노르웨이', '벨기에', '오스트리아', \
                           '스페인', '그리스', '포르투갈', \
                           '핀란드', '아일랜드', '우크라이나', \
                           '러시아', '크로아티아', '루마니아', \
                           '불가리아', '폴란드','구주 기타']):
        continents.append('유럽')
    elif nation_name in (['오스트레일리아', '뉴질랜드', '대양주 기타']):
        continents.append('오세아니아')
    elif nation_name in (['남아프리카공화국', '아프리카 기타']):
        continents.append('아프리카')
    elif nation_name in (['국적미상']):
        continents.append('기타대륙')
    elif nation_name in (['교포']):
        continents.append('기타대륙')
continents
```

```python
# 관광객 비율(%)
kto2['관광객 비율(%)'] = round(kto2['관광']/kto2['계']*100, 1)
kto2.head()
```

```python
# 관광객 비율 내림차순 출력
kto2.sort_values(by='관광객 비율(%)', ascending=False).head(15)
```

```python
# '계' 내림차순 출력
kto2.sort_values(by='계', ascending=False).head(5)
```

```python
# pivot_table() 함수 활용하기, 대륙 컬럼을그룹화하여 관광객비율 컬럼의 평균 산출
# index가 그룹
# kto2.pivot_table(values='관광객 비율(%)', index='대륙') # 대륙별 평균이 default로 적용
# kto2.pivot_table(index='대륙', values='관광객 비율(%)') # 순서 변경
kto2.pivot_table(index='대륙', 
                 values='관광객 비율(%)',
                 aggfunc='mean' # 평균 적용
                )
```

```python
# pivot_table() 함수 활용하기, 대륙 컬럼을그룹화하여 관광 컬럼의 합 산출
kto2.pivot_table(index='대륙', 
                 values='관광',
                 aggfunc='sum' # 평균 적용
                )
```

```python
# 중국 국적만 필터링
condition = (kto2.국적 == '중국')
# condition = (kto2['국적'] == '중국') # 위와 같다
kto2[condition]
```

```python
# 2019년 1월 우리나라 방문하는 전체 외국인 관광객 숫자 구하기
tour_sum= sum(kto2['관광'])
tour_sum
```

```python
# 칼럼 삭제
# del kto2['전체비율']
# kto2.drop(['전체비율'], axis='columns', inplace=True)
```

```python
# 관광 / 전체 관광객 * 100 = 전체비율(%)
# 전체비율 칼럼 기준으로 내림차순 정렬하기
kto2['전체비율(%)'] = round(kto2['관광']/tour_sum*100, 1)
kto2.sort_values(by='전체비율(%)', ascending=False).head()
```
