```python
# 하나의 국가명을 전달 받아 대륙명을 리턴하는 함수
def getContinent(nation_name):
    continents = ''
    
    if "'일본', '대만', '홍콩', '마카오', '태국', \
        '말레이시아', '필리핀', '인도네시아',\
        '싱가포르', '미얀마', '베트남', '인도',\
        '스리랑카', '파키스탄', '방글라데시',\
        '캄보디아', '몽골', '중국', '이란', \
        '이스라엘', '터키', '우즈베키스탄', \
        '카자흐스탄', 'GCC', '아시아 기타'".find(nation_name) >= 0: 
        continents='아시아'
    elif "'미국', '캐나다', '멕시코', '브라질', \
         '미주 기타'".find(nation_name) >= 0: 
        continents='아메리카'
    elif "'영국', '독일', '프랑스', '네덜란드','스웨덴', '스위스', \
          '이탈리아', '덴마크', '노르웨이', '벨기에', '오스트리아',\
          '스페인', '그리스', '포르투갈', '핀란드', '아일랜드',\
          '우크라이나', '러시아', '크로아티아', '루마니아',\
          '불가리아', '폴란드', '구주 기타'".find(nation_name) >= 0:
        continents='유럽'
    elif "'오스트레일리아', '뉴질랜드', '대양주 기타'".find(nation_name) >= 0:
        continents='오세아니아'         
    elif "'남아프리카공화국', '아프리카 기타'".find(nation_name) >= 0:
        continents='아프리카'
    elif "'국적미상'".find(nation_name) >= 0:
        continents='기타대륙'         
    elif "'교포'".find(nation_name) >= 0:
        continents='교포'
    else:
        print(nation_name + ' 발견 안됨')
        
    return continents
```

```python
def getContinents(df):
    continents = [] # 대륙명 저장
    for i in range(len(df)):
        nation_name = df.loc[i][0]
        # print(nation_name) # 국가명
        if "'일본', '대만', '홍콩', '마카오', '태국', \
            '말레이시아', '필리핀', '인도네시아',\
            '싱가포르', '미얀마', '베트남', '인도',\
            '스리랑카', '파키스탄', '방글라데시',\
            '캄보디아', '몽골', '중국', '이란', \
            '이스라엘', '터키', '우즈베키스탄', \
            '카자흐스탄', 'GCC', '아시아 기타'".find(nation_name) >= 0: 
            continents.append('아시아')
        elif "'미국', '캐나다', '멕시코', '브라질', \
             '미주 기타'".find(nation_name) >= 0: 
            continents.append('아메리카') 
        elif "'영국', '독일', '프랑스', '네덜란드','스웨덴', '스위스', \
              '이탈리아', '덴마크', '노르웨이', '벨기에', '오스트리아',\
              '스페인', '그리스', '포르투갈', '핀란드', '아일랜드',\
              '우크라이나', '러시아', '크로아티아', '루마니아',\
              '불가리아', '폴란드', '구주 기타'".find(nation_name) >= 0:
            continents.append('유럽') 
        elif "'오스트레일리아', '뉴질랜드', '대양주 기타'".find(nation_name) >= 0:
            continents.append('오세아니아')         
        elif "'남아프리카공화국', '아프리카 기타'".find(nation_name) >= 0:
            continents.append('아프리카')  
        elif "'국적미상'".find(nation_name) >= 0:
            continents.append('기타대륙')          
        elif "'교포'".find(nation_name) >= 0:
            continents.append('교포')
    return continents
# getContinents(df)
```

```python
# 년월을 입력 받아 DataFrame을 리턴하는 함수
def create_data(yy, mm):
    df = pd.DataFrame() # return 할 객체
    mm = str(mm).zfill(2) # 1 -> 01
    file_path = './data/kto_{0}{1}.xlsx'.format(yy, mm)
    if os.path.exists(file_path):
        df = pd.read_excel(file_path, header=1, skipfooter=4, usecols='A:G')
        
        # 기준년월 칼럼 추가
        df['기준년월'] = '{0}-{1}'.format(yy, mm)
        
        # 순수 국가명만 추출
        continents = ['아시아주', '미주', '구주', '대양주', \
                      '아프리카주', '기타대륙', '교포소계']
        df = df[df.국적.isin(continents) == False]
        df = df.reset_index(drop=True) # 다시 인덱스를 0부터 설정
        
        # 대륙 칼럼 추가
        df['대륙'] = getContinents(df)
        
        # 관광객 비율(%)
        df['관광객 비율(%)'] = round(df['관광']/df['계']*100, 1)
        
        # 관광 / 전체 관광객 * 100 = 전체비율(%)
        tour_sum= sum(df['관광'])
        df['전체비율(%)'] = round(df['관광']/tour_sum*100, 1)
    return df
```

```python
df = create_data(2020, 4)
df
```

```python
kto = pd.DataFrame()
for yy in range(2010, 2021): # 2010~2020
    for mm in range(1, 13): # 01~12
        rdf = create_data(yy, mm)
        kto = kto.append(rdf, ignore_index=True) # 원본 인덱스를 유지하지 않고, 계속 증가시킴
```
