```python
age_sum = [0] * 6
f = open('./sales.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
header = next(data)
for row in data:
    for i in range(1, 7): # 1 ~ 6
        # age_sum[i-1]: index 0부터 시작
        age_sum[i-1] = age_sum[i-1] + int(row[i]) # 연령별 합계

plt.figure(figsize=(10, 7))
area = age_sum # pie plot area
label = header[1:]
color = ['darkorange', 'gold', 'yellowgreen', 'greenyellow',
         'lightcoral', 'lightskyblue']
plt.pie(area, colors=color, labels=label, autopct='%.1f%%',
        startangle=0, shadow=True, textprops={'fontsize': 14})
plt.title('눈꽃 Resort 예약 분포 현황')
plt.legend()
plt.axis('equal')

# subway_img 폴더는 사전에 생성할것
# if os.path.exists('./subway_img/') == False:
#     os.mkdir('./subway_img/')
    
plt.savefig('./sales.png') # 파일 생성
plt.show()

f.close()
```
