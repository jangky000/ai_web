```python
# 서울 기준 매년 8월의 최고 기온만 추출하여 plot 출력
f = open('./seoul.csv', 'r', encoding='utf-8') # r: read 읽기모드
data = csv.reader(f, delimiter=',') # delimiter: 구분자
header = next(data) # 현재 라인을 읽고, 파일 지시자를 다음 라인으로 이동(첫줄읽음)
result = []


for row in data:
    if row[-1] != '': # 데이터가 있는 경우
        month = row[0].split('-')[1] # 월만 추출
        if month == '08':
            result.append(float(row[-1]))

f.close()
```

```python
x = [year for year in range(1907, 1927, 1)]
y = result[0:20]

plt.figure(figsize=(8,3))
plt.plot(x,y,c='red', label='온도')
plt.title('1907~1926년 8월의 온도분포')
plt.legend()
plt.grid(True)
plt.xticks(x)
plt.show()
```

```python
# 매년 12월 25일의 최고 기온만 추출하여 plot 출력
f = open('./seoul.csv', 'r', encoding='utf-8') # r: read 읽기모드
data = csv.reader(f, delimiter=',') # delimiter: 구분자
header = next(data) # 현재 라인을 읽고, 파일 지시자를 다음 라인으로 이동(첫줄읽음)
result = []


for row in data:
    if row[-1] != '': # 데이터가 있는 경우
        month = row[0].split('-')[1] # 월만 추출
        day = row[0].split('-')[2] # 일만 추출
        if month == '12' and day == '25':
            result.append(float(row[-1]))

plt.plot(result,c='blue', label='온도')
plt.title('매년 크리스마스의 온도분포')
plt.legend()
plt.grid(True)
plt.show()

f.close()
```
